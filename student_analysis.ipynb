## Step 1: Load Dataset
We load the student performance dataset using pandas.

```python
import pandas as pd

df = pd.read_csv("student-mat.csv", sep=";")
df.head()
```

## Step 2: Data Cleaning and Exploration
We check for missing values, duplicates, and understand the dataset.

```python
# Check missing values
df.isnull().sum()
```

```python
# Remove duplicates
df = df.drop_duplicates()
```

```python
# Check shape
df.shape
```

```python
# Check data types
df.dtypes
```

## Step 3: Data Analysis
We analyze student performance using different factors.

### Average Final Grade

We calculate the average of the final grades (G3) to understand overall student performance.

```python
df["G3"].mean()
```

### Students Scoring Above 15

We count the number of students who scored above 15 to identify high-performing students.

```python
len(df[df["G3"] > 15])
```

### Study Time vs Performance

We calculate the correlation between study time and final grades to see if study time impacts performance.

```python
df["studytime"].corr(df["G3"])
```

### Gender-wise Performance

We compare the average final grades of male and female students to observe performance differences.

```python
df.groupby("sex")["G3"].mean()
```

## Step 4: Data Visualization
We visualize patterns in the dataset.

### Distribution of Final Grades

This histogram shows how student grades are distributed.

```python
import matplotlib.pyplot as plt

plt.hist(df["G3"])
plt.title("Distribution of Final Grades")
plt.xlabel("Grades")
plt.ylabel("Frequency")
plt.show()
```

### Study Time vs Grades

This scatter plot shows the relationship between study time and final grades.

```python
plt.scatter(df["studytime"], df["G3"])
plt.title("Study Time vs Grades")
plt.xlabel("Study Time")
plt.ylabel("Grades")
plt.show()
```

### Average Score by Gender

This bar chart compares the average grades of male and female students.

```python
df.groupby("sex")["G3"].mean().plot(kind="bar")
plt.title("Average Score by Gender")
plt.show()
```

## Conclusion

- The dataset shows overall student performance through final grades.
- A portion of students scored above 15, indicating strong performers.
- Study time has a weak correlation with grades, meaning other factors also affect performance.
- Gender comparison shows only slight differences in performance.
- Overall, student performance depends on multiple factors beyond study time alone.
